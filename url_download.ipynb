```python
import pandas as pd
```

```python
df = pd.read_csv('data/raw/multimedia.csv')
df.head()
```

```python
df[['_id', 'identifier']].sample(n=10)
```

```python
urls = df['identifier'].sample(n=1000)
```

```python
import requests
import io
from PIL import Image
from tqdm import tqdm

output_size = (512, 512)

for url in tqdm(urls.items()):
    r = requests.get(url[1])
    image_bytes = io.BytesIO(r.content)
    im = Image.open(image_bytes)
    im.thumbnail(output_size)
    im.save('data/download/' +str(url[0]) + '.jpg')
```
